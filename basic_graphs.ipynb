# Basic graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_nodes_from([2, 3, 4, 5])
```

```python
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(3, 4)
G.add_edge(3, 5)
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
G.nodes
```

```python
G.edges
```

```python
G.degree[1]
```

```python
list(G.adj[1])
```

```python
nx.draw(G, with_labels=True)
```
